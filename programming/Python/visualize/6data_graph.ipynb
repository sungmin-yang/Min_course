```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def save_graph(title, tfidf, fasttext, word2vec_cnn, awd_lstm, last_instance):
    assert isinstance(title, str)
    target_list = []
    instance = ['10', '50', '100', '200', '500', '1000', '2000', last_instance]

    df_input1 = np.stack([tfidf, fasttext, word2vec_cnn]).T
    df_input2 = np.stack([awd_lstm]).T
    # pd_input = np.stack([test_hist1, test_hist2]).T # or transpose()


    df1 = pd.DataFrame(df_input1, index=instance, columns=['tfidf', 'fasttext', 'word2vec_CNN'])
    df2 = pd.DataFrame(df_input2, index=instance, columns=['AWD-LSTM'])

    plt.figure()
    # plt.plot(df1, markersize=12, linewidth=1 )
    # plt.plot(df2, markerfacecolor='blue', marker='o', markersize=5, color='blue', linewidth=4 )
    ax = df1.plot(title=title, markersize=50, linestyle='--' )
    ax = df2.plot(title=title, color='red', markersize=5, marker='o', ax=ax)

    ax.set_xlabel("Instance per class")
    ax.set_ylabel("Accuracy")
    plt.savefig('{}.jpg'.format(title))
    
    print("Successfully saved file {}".format(title))
    
    return None


tfidf        = np.array([48.78, 71.15, 76.32, 79.82 , 83.35 , 85.29 , 86.19 ,90.36 ])
fasttext     = np.array([30, 54, 63, 71.8, 79.59, 82.86, 85.86, 91.38])
word2vec_cnn = np.array([ 45.19, 76.92, 79.93, 80.94, 85.43, 87.29, 88.54, 92.07])
awd_lstm     = np.array([69.57, 85.8, 88.42, 89.14, 90.42, 90.59, 91.4, 91.17])

save_graph("Agnews", tfidf, fasttext, word2vec_cnn, awd_lstm, 30000)
```
