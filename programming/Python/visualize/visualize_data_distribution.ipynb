```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
trip_dist = {'1':9496, '2':9821, '3':18462, '4':44700, '5':59324}
ama_dist  = {'cameras': 75502, 'labtops': 6432, 'mobilephone':29874, 
            'tablets':19831, 'tv':37278, 'video_survellance':9369}
```

```python
def get_instance_class_from_dist(data_dist):
    instance, class_name = [], []
    for c, i in data_dist.items():
        instance.append(i)
        class_name.append(c)
    return instance, class_name

instance, class_name = get_instance_class_from_dist(trip_dist)
df_trip = pd.DataFrame(instance, index=class_name, columns=['Tripadvisor'])

instance, class_name = get_instance_class_from_dist(ama_dist)
df_ama = pd.DataFrame(instance, index=class_name, columns=['Amazon_six'])
```

```python
# fig, axes = plt.subplots(nrows=1, ncols=2)

# df_trip.plot(ax=axes[0], kind='pie', subplots=True, figsize=(8,8))
# df_ama.plot(ax=axes[1], kind='pie', subplots=True, figsize=(8,8))
df_ama.plot(kind='pie', subplots=True, figsize=(8,8), autopct='%1.0f%%')
df_trip.plot(kind='pie', subplots=True, figsize=(8,8), autopct='%1.0f%%')
```

```python
agnews_dist = {'1':1, '2':1, '3':1, '4':1}
instance, class_name = get_instance_class_from_dist(agnews_dist)
df_agnews = pd.DataFrame(instance, index=class_name, columns=['Agnews'])

db_dist = {'1':1, '2':1, '3':1, '4':1, '5':1, '6':1, '7':1, '8':1, '9':1, '10':1, '11':1, '12':1, '13':1, '14':1}
instance, class_name = get_instance_class_from_dist(db_dist)
df_db = pd.DataFrame(instance, index=class_name, columns=['DBpedia'])

yahoo_dist = {'1':1, '2':1, '3':1, '4':1, '5':1, '6':1, '7':1, '8':1, '9':1, '10':1}
instance, class_name = get_instance_class_from_dist(yahoo_dist)
df_yahoo = pd.DataFrame(instance, index=class_name, columns=['Yahoo answer'])

yelp_dist = {'1':1, '2':1, '3':1, '4':1, '5':1}
instance, class_name = get_instance_class_from_dist(yelp_dist)
df_yelp = pd.DataFrame(instance, index=class_name, columns=['Yelp review'])

plt.figure()
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10,10))
df_agnews.plot(kind='pie', subplots=True, figsize=(4,4), autopct='%1.0f%%', ax=axes[0,0])
df_db.plot(kind='pie', subplots=True, figsize=(4,4), autopct='%1.0f%%', ax=axes[0,1])
df_yahoo.plot(kind='pie', subplots=True, figsize=(4,4), autopct='%1.0f%%', ax=axes[1,0])
df_yelp.plot(kind='pie', subplots=True, figsize=(4,4), autopct='%1.0f%%', ax=axes[1,1])
plt.savefig('balanced_class.jpg')
```
